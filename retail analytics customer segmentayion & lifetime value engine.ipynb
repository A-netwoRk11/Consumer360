```python
import pandas as pd

df = pd.read_excel("Consumer360.xlsx")
df
```

```python
rfm_df = df[['customer_id', 'order_date', 'order_value']]
rfm_df 
```

```python
rfm_df['order_date'] = pd.to_datetime(rfm_df['order_date'])
rfm_df['order_date']
```

```python
reference_date = rfm_df['order_date'].max() + pd.Timedelta(days=1)
reference_date
```

```python
rfm = rfm_df.groupby('customer_id').agg({
    'order_date': lambda x: (reference_date - x.max()).days,
    'customer_id': 'count',
    'order_value': 'sum'
})
rfm 
```

```python
rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.columns
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['R_Score']
```

```python
rfm['F_Score'] = pd.cut(
    rfm['Frequency'].rank(method='first'),
    5,
    labels=[1,2,3,4,5]
)
rfm['F_Score']
```

```python
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
rfm['M_Score']
```

```python
rfm['RFM_Score'] = (
    rfm['R_Score'].astype(int) +
    rfm['F_Score'].astype(int) +
    rfm['M_Score'].astype(int)
)
rfm['RFM_Score']
```

```python
def segment_customer(score):
    if score >= 13:
        return "Whales"
    elif score >= 9:
        return "Loyal Customers"
    elif score >= 6:
        return "Potential"
    else:
        return "Churn Risk"

rfm['Segment'] = rfm['RFM_Score'].apply(segment_customer)
rfm['Segment'] 
```
